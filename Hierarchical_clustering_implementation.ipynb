```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
data = datasets.load_iris()
x = data.data
y = data.target
```

```python
x
```

```python
y
```

```python
# scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# reducing dimensions applying pca

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca
```

```python
pca_scaled = pca.fit_transform(x_scaled)
pca_scaled
```

```python
plt.scatter(pca_scaled[:, 0],pca_scaled[:, 1],c =y)
```

```python
# hierarchical clustering
# agglomerative clustering >> make dendogram
import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20, 7))
plt.title('Dendrogram')

sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
plt.xlabel('sample data')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
# build the model

from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[: ,0],pca_scaled[:,1],c =cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score

silhoutte_coef = []
for k in range(2,11):
  agglo = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
  agglo.fit(x_scaled)
  score = silhouette_score(x_scaled, agglo.labels_)
  silhoutte_coef.append(score)
```
